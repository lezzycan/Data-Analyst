```python
#multi-indexing
# Index Levels
outside = ['G1', 'G1', 'G1', 'G2', 'G2', 'G2']
inside = [1,2,3,1,2,3]
hier_index = list(zip(outside,inside))
```

```python
print("\nTuple pairs after the zip and list command\n", '-'*45, sep='')
print(hier_index)
```

```python
hier_index = pd.MultiIndex.from_tuples(hier_index)
print("\nIndex hierachy\n", '-'*25, sep = '')
print(hier_index)
```

```python
print("\nCreating DataFrame with multi-index\n", '-'*37, sep = '')
#np.random.seed(101)
df1 = pd.DataFrame(data = np.round(rn(6,3)), index = hier_index, columns = ['A', 'B','C'])
print(df1)
```

```python
L1 = ['A','A','A','B','B','B','C','C','C']
L2 = [1,2,3,1,2,3,1,2,3]
L3 = [1,2,3,4,5,6,7,8,8,9]
```

```python
multi_index = list(zip(L1,L2,L3))
print(multi_index)
```

```python
multi_index = pd.MultiIndex.from_tuples(multi_index)
print(multi_index)
```
